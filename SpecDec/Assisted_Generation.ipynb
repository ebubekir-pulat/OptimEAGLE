```python
# Code From: https://huggingface.co/blog/assisted-generation

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

prompt = "My hands hurt after"
checkpoint = "roneneldan/TinyStories-8M"
assistant_checkpoint = "roneneldan/TinyStories-1M"
device = "cuda" if torch.cuda.is_available() else "cpu"

tokenizer = AutoTokenizer.from_pretrained(checkpoint)
inputs = tokenizer(prompt, return_tensors="pt").to(device)

model = AutoModelForCausalLM.from_pretrained(checkpoint).to(device)
assistant_model = AutoModelForCausalLM.from_pretrained(assistant_checkpoint).to(device)
```

```python
import time

start = time.perf_counter_ns()

# 2 Code Lines Below From: https://huggingface.co/blog/assisted-generation
outputs = model.generate(**inputs, assistant_model=assistant_model)
print(tokenizer.batch_decode(outputs, skip_special_tokens=True))

finish = time.perf_counter_ns()
elapsed = finish - start
print("Wall Clock Time (ns): ", elapsed)
```
